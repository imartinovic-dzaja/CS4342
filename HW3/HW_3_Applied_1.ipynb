```python
import seaborn as sns

import sklearn.linear_model as skl_lm
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, classification_report, precision_score
from sklearn import preprocessing
from sklearn import neighbors

import statsmodels.api as sm
import statsmodels.formula.api as smf



predictors = data.drop(['Direction', 'Year', 'Today'], axis=1)
predictors = sm.add_constant(predictors)
est = sm.Logit(data['Direction'].ravel(), predictors).fit()
est.summary2().tables[1]

```

```python
predictions = est.predict(predictors).astype(int);

cm_df = pd.DataFrame(confusion_matrix(data['Direction'], predictions).T, index=["Down", "Up"], columns=["Down", "Up"])
cm_df.index.name = 'Predicted'
cm_df.columns.name = 'True'
print(cm_df)  
print(classification_report(data['Direction'], predictions, digits=3))
```

```python
trainingData = data.loc[data['Year'] <= 2008]
testData = data.loc[data['Year'] > 2008]

trainingPredictors = trainingData['Lag2']
trainingPredictors = sm.add_constant(trainingPredictors)
est = sm.Logit(trainingData['Direction'].ravel(), trainingPredictors).fit()

testPredictors = testData['Lag2']
testPredictors = sm.add_constant(testPredictors)
testPredictions = est.predict(testPredictors).astype(int);

cm_df = pd.DataFrame(confusion_matrix(testData['Direction'], testPredictions).T, index=["Down", "Up"], columns=["Down", "Up"])
cm_df.index.name = 'Predicted'
cm_df.columns.name = 'True'
print(cm_df)  
print(classification_report(testData['Direction'], testPredictions, digits=3))
```

```python
lda = LinearDiscriminantAnalysis(solver='svd')
testPredictions = lda.fit(trainingPredictors, trainingData['Direction']).predict(testPredictors)

cm_df = pd.DataFrame(confusion_matrix(testData['Direction'], testPredictions).T, index=["Down", "Up"], columns=["Down", "Up"])
cm_df.index.name = 'Predicted'
cm_df.columns.name = 'True'
print(cm_df)  
print(classification_report(testData['Direction'], testPredictions, digits=3))
```

```python
qda = QuadraticDiscriminantAnalysis()
testPredictions = qda.fit(trainingPredictors, trainingData['Direction']).predict(testPredictors)

cm_df = pd.DataFrame(confusion_matrix(testData['Direction'], testPredictions).T, index=["Down", "Up"], columns=["Down", "Up"])
cm_df.index.name = 'Predicted'
cm_df.columns.name = 'True'
print(cm_df)  
print(classification_report(testData['Direction'], testPredictions, digits=3))
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=1)
testPredictions = knn.fit(trainingPredictors, trainingData['Direction']).predict(testPredictors)
cm_df = pd.DataFrame(confusion_matrix(testData['Direction'], testPredictions).T, index=["Down", "Up"], columns=["Down", "Up"])
cm_df.index.name = 'Predicted'
cm_df.columns.name = 'True'
print(cm_df)  
print(classification_report(testData['Direction'], testPredictions, digits=3))
```
