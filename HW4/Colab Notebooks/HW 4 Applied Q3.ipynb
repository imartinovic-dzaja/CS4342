```python
import numpy as np
import pandas as pd
import math

data = pd.read_csv('Boston.csv')

medv = data['medv']

print(medv.mean())
print(medv.std() / math.sqrt(data.shape[0]))


```

```python
from sklearn.utils import resample

stdDevSum = 0
n = 10000
for i in range(n):
  X = resample(medv, random_state=i)
  stdDevSum = stdDevSum + X.std()

stdDev = stdDevSum / n
stdDev = stdDev / math.sqrt(X.size)

print(stdDev)
```

```python
print([medv.mean() - 2 * stdDev, medv.mean() + 2*stdDev])
```
